```python
import pandas as pd
import numpy as np
import seaborn as sns


# Import the libraries we'll use to create graphs
from matplotlib import pyplot as plt

# This command makes sure the graphs show in this window, not a new one
%matplotlib inline 
```

```python
x=data['BMI']
plt.xlabel('BMI Value')
plt.title('Average BMI')
sns.distplot(x)
plt.show()
```

```python
x=data['BloodPressure']
y=data['Age']
plt.xlabel('Age')
plt.ylabel('Blood Pressue')
plt.title('Relationship Between Age and Blood Pressure')
sns.scatterplot(x,y)
plt.show()
```

```python
BP=data.BloodPressure
print(max(BP))
print(min(BP))
```

```python
data['BloodPressure']=data['BloodPressure'].replace({0:np.nan})
```

```python
print (min(data.BloodPressure))
```

```python
x=data['Age']
y=data['BloodPressure']
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.title('Relationship Between Age and Blood Pressure')
sns.regplot(x,y,scatter_kws={"color":"blue"}, line_kws={"color":"red"})

plt.show()
```

```python
x=data['Pregnancies']
data.Glucose=data.Glucose.replace({0:np.nan})
y=data['Glucose']
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.title('Relationship Between Pregancy and Glucose')
sns.boxplot(x,y)
plt.show()
```

```python
x=data['Age']
data.Glucose=data.Glucose.replace({0:np.nan})
y=data['Glucose']
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.title('Relationship Between Age and Blood Pressure')
sns.kdeplot(x,y) #non-parametric way to estimate the probability density 
                    #function of a continuous random variable
plt.show()
```

```python
x=data['Age']
data.Glucose=data.Glucose.replace({0:np.nan})
y=data['Glucose']
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.title('Relationship Between Age and Blood Pressure')
sns.regplot(x,y)
plt.show()
```

```python
x=data['Age']
data.Insulin=data.Insulin.replace({0:np.nan})
y=data['Insulin']
plt.xlabel('Age')
plt.ylabel('Insulin')
plt.title('Relationship Between Age and Insulin')
sns.regplot(x,y)
plt.show()
```
